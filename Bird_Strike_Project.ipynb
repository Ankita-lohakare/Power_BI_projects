# Handling Missing Values:

```python
df2 = df2.fillna({'Airport: Name': df2['Airport: Name'].mode()[0]})
```

```python
df2 = df2.fillna({'Altitude bin(>1000)': df2['Altitude bin(>1000)'].median()})
```

```python
df2 = df2.fillna({'Effect: Impact to flight': df2['Effect: Impact to flight'].mode()[0]})
```

```python
df2 = df2.fillna({'FlightDate': df2['FlightDate'].mode()[0]})
```

```python
df2 = df2.fillna({'Aircraft: Airline/Operator': df2['Aircraft: Airline/Operator'].mode()[0]})
```

```python
df2 = df2.fillna({'Origin State': df2['Origin State'].mode()[0]})
```

```python
df2 = df2.fillna({'When: Phase of flight': df2['When: Phase of flight'].mode()[0]})
```

```python
df2 = df2.fillna({'Pilot warned of birds or wildlife?': df2['Pilot warned of birds or wildlife?'].mode()[0]})
```

```python
df2 = df2.fillna({'Wildlife: Size': df2['Wildlife: Size'].mode()[0]})
```

```python
df2 = df2.fillna({'Feet above ground': df2['Feet above ground'].mean()})
```

```python
df2.isnull().sum()
```

```python
df2.to_csv('Bird_strike_df2.csv') # saving in csv
```
